```python
def printSolution(board):
    for i in range(x):
        for j in range(x):
            print(" Q " if board[i][j] == 1 else " . ", end="")
        print()
            
    

def is_safe(board,row,col):
    for i in range(row):
        if board[i][col] == 1:
            return False
        
    for i,j in zip(range(row, -1, -1) , range(col, -1,-1)):
        if board[i][j] == 1:
            return False
    
    for i,j in zip(range(row, x, 1) , range(col, -1,-1)):
        if board[i][j] == 1:
            return False
    
    for i, j in zip(range(row, -1, -1), range(col , x ,1)):
        if board[i][j] == 1:
            return False
    
    for i, j in zip(range(row, x, 1), range(col , x ,1)):
        if board[i][j] == 1:
            return False

    return True

def solve_queen(board, row):
    if row == len(board):
        printSolution(board)
        return True

    found_solution = False

    for col in range(len(board)):
        if is_safe(board, row, col):
            board[row][col] = 1
            found_solution |= solve_queen(board, row + 1)
            board[row][col] = 0

    return found_solution

def solve_all_queens(x):
    board = [[0 for _ in range(x)] for _ in range(x)]
    solutions_found = solve_queen(board, 0)
    
    if not solutions_found:
        print("No solutions found")

if __name__ == "__main__":
    x = int(input("Enter the board size: "))
    solve_all_queens(x)
```

```python

#Using Adjacency List

graph = {}

def add_edges(graph, u,v, wt):
    if u not in graph:
        graph[u] = []
    if v not in graph:
        graph[v] = []
    graph[u].append((v,wt))
    graph[v].append((u,wt))
    
    
add_edges(graph, 0, 1, 5)
add_edges(graph, 1,2,6)
add_edges(graph, 0,2,3)

print(f"Neighbours of 0: {graph[0]}", ) 
print(f"Neighbours of 2: {graph[2]}", ) 
```

```python
# Using edge-wise

graph = []

def add_EDGE(graph, u, v, wt):
    graph.append((u,v, wt))
    graph.append((v,u,wt))
    
    
add_EDGE(graph, 0,1,6)
add_EDGE(graph, 0,2, 5)
add_EDGE(graph, 1,2, 5)

print(graph)
```
