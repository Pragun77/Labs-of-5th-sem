```python
#Bubble sort Algorithm
import matplotlib.pyplot as plt
import random
import time
import csv

def bubble_sort(arr:list):
    for i in range(len(arr)):
        swapped = False
        for j  in range(len(arr)-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        if swapped == False:
            break
        
def create_list(length: int):
    '''
    Create a list of random whole numbers less than 100 of length n.

    Parameters:
        length (int): The length of the list to be created.
        dataRange (list[int]): Range of data with index 0 as minimum value and index 1 as maximum value.

    Returns:
        list[int]: A list with random generated numbers.
    '''
    datarange = [1,100]
    numbers = []
    for i in range(length):
        numbers.append(random.randint(datarange[0], datarange[1]))
    return numbers

def generate_graph(lenarray: list, timeArray: list):
    '''
    Plot a graph from the given csv .FIle

    Parameter:
    csvFile (str): The system path to the file with input values.
    '''

    # Extract x and y values from the data
    x_values = lenarray
    y_values = timeArray

    # Create a line plot
    plt.plot(x_values, y_values, label='Data Points', color='blue', marker='.')

    # Add labels and title
    plt.xlabel('Length of Array')
    plt.ylabel('Time of Execution')
    plt.title('Execution TIme v/s Array Length')

    # Add gridlines
    plt.grid()

    # Show the plot
    plt.legend()
    plt.show()

def Analysis(arr:list):
    t= []
    start_time = time.time()
    y = bubble_sort(arr=arr)
    end_time = time.time()
    exec_time = end_time - start_time
    t.append(exec_time)
    return t
    

if __name__ == '__main__':
    lengths = [100,500,1000,2000,4000,5000,10000]
    execution_times= []
    for length in lengths:
        a= create_list(length)
        execution_time = Analysis(a)
        execution_times.append(execution_time)
        
    print(execution_times)
    with open("C:\\Users\\pragu\\OneDrive\\Desktop\\Book1.csv", 'w', newline='') as csvFile:
        csv_writer = csv.writer(csvFile)
        # Adding headings to the CSV file
        csv_writer.writerow(['Run', 'Array Length', 'Execution Time'])
        for i in range(len(lengths)):
            csv_writer.writerow([i, lengths[i], execution_times[i]])
    generate_graph(lengths , execution_times)
```

```python
#quick sort Algorithm
import matplotlib.pyplot as plt
import random
import time
import csv

def partition(array:list, low, high):
    pivot = array[high]
    i = low-1
    
    for j in range(low,high):
        if array[j] <= pivot:
            i = i+1
            (array[i],array[j]) =( array[j] , array[i])
    (array[i+1], array[high]) = (array[high], array[i+1])
    return i+1
            

def quick_sort(arr:list, low ,high):
    if low<high:
        pi = partition(arr, low,high)
        quick_sort(arr, low, pi-1)
        quick_sort(arr, pi+1, high)
        
def create_list(length: int):
    '''
    Create a list of random whole numbers less than 100 of length n.

    Parameters:
        length (int): The length of the list to be created.
        dataRange (list[int]): Range of data with index 0 as minimum value and index 1 as maximum value.

    Returns:
        list[int]: A list with random generated numbers.
    '''
    datarange = [1,100]
    numbers = []
    for i in range(length):
        numbers.append(random.randint(datarange[0], datarange[1]))
    return numbers

def generate_graph(lenarray: list, timeArray: list):
    '''
    Plot a graph from the given csv .FIle

    Parameter:
    csvFile (str): The system path to the file with input values.
    '''

    # Extract x and y values from the data
    x_values = lenarray
    y_values = timeArray

    # Create a line plot
    plt.plot(x_values, y_values, label='Data Points', color='blue', marker='.')

    # Add labels and title
    plt.xlabel('Length of Array')
    plt.ylabel('Time of Execution')
    plt.title('Execution TIme v/s Array Length')

    # Add gridlines
    plt.grid()

    # Show the plot
    plt.legend()
    plt.show()

def Analysis(arr:list):
    t= []
    start_time = time.time()
    quick_sort(arr=arr, low=0, high= len(arr) -1)
    end_time = time.time()
    exec_time = end_time - start_time
    t.append(exec_time)
    return t
    

if __name__ == '__main__':
    lengths = [100,500,1000,2000,4000,5000,8000,10000,12000,14000,16000,20000]
    execution_times= []
    
    for length in lengths:
        a= create_list(length)
        execution_time = Analysis(a)
        execution_times.append(execution_time)

    print(execution_times)
    with open("C:\\Users\\pragu\\OneDrive\\Desktop\\Book1.csv", 'w', newline='') as csvFile:
        csv_writer = csv.writer(csvFile)
        # Adding headings to the CSV file
        csv_writer.writerow(['Index', 'Array Length', 'Execution Time'])
        for i in range(len(lengths)):
            csv_writer.writerow([i, lengths[i], execution_times[i]])

    generate_graph(lengths , execution_times)
```
