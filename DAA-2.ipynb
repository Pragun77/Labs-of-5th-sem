```python
import random
import csv
import matplotlib.pyplot as plt
import timeit

def linear_search(arr,n, target):
    for i in range(0,n):
        if (arr[i] == target):
            return i
    return -1

def create_list(length):
    datarange = [1, 100]
    numbers = []
    for _ in range(length):
        numbers.append(random.randint(datarange[0], datarange[1]))
    return numbers

def generate_graph(lenarray, timeArray_best, time_arr_worst, time_arr_avg):
    x_values = lenarray
    y_best_values = timeArray_best
    y_worst_values = time_arr_worst
    y_avg_values = time_arr_avg

    plt.plot(x_values, y_best_values, label='Best Case', color='green', marker='.')
    plt.plot(x_values, y_worst_values, label='Worst Case', color='red', marker='.')
    plt.plot(x_values, y_avg_values, label='Average Case', color='blue', marker='.')

    plt.xlabel('Length of Array')
    plt.ylabel('Time of Execution')
    plt.title('Execution Time vs. Array Length')
    plt.grid()
    plt.legend()
    plt.show()

def analysis_best_case(arr):

    return timeit.timeit(lambda:linear_search(arr, len(arr), arr[0]), number=1000)

def analysis_worst_case(arr):

    return timeit.timeit(lambda:linear_search(arr, len(arr), arr[len(arr)- 1]), number=1000)

def analysis_avg_case(arr):

    return timeit.timeit(lambda:linear_search(arr, len(arr), arr[random.randint(0,len(arr)-1)]) , number=1000)

if __name__ == '__main__':
    lengths = [ 10000 , 20000, 40000, 80000, 150000, 400000, 1000000]
    best_case = []
    worst_case = []
    avg_case = []

    for length in lengths:
        a = create_list(length)

        best = analysis_best_case(a)
        best_case.append(best)  # Extend the list with the results
        worst = analysis_worst_case(a)
        worst_case.append(worst)
        avg = analysis_avg_case(a)
        avg_case.append(avg)  # Extend the list with the results

    print(best_case)
    print(avg_case)
    print(worst_case)

    # print()
    with open("C:\\Users\\pragu\\OneDrive\\Desktop\\Book1.csv", 'w', newline='') as csvFile:
        csv_writer = csv.writer(csvFile)
        # Adding headings to the CSV file
        csv_writer.writerow(['Run', 'Array Length', 'Execution Time Best', 'Execution Time worst', 'Execution Time Avg'])
        for i in range(len(lengths)):
            csv_writer.writerow([i, lengths[i], best_case[i], worst_case[i], avg_case[i]])
    generate_graph(lengths , best_case, worst_case, avg_case)


    
        
```

```python
import random
import csv
import matplotlib.pyplot as plt
import timeit

def binary_search(arr, x):
    low = 0
    high = len(arr) - 1
    mid = 0
 
    while low <= high:
 
        mid = (high + low) // 2
 
        # If x is greater, ignore left half
        if arr[mid] < x:
            low = mid + 1
 
        # If x is smaller, ignore right half
        elif arr[mid] > x:
            high = mid - 1
 
        # means x is present at mid
        else:
            return mid
 
    # If we reach here, then the element was not present
    return -1

def create_list(length):
    datarange = [1, 100]
    numbers = []
    for _ in range(length):
        numbers.append(random.randint(datarange[0], datarange[1]))
    return numbers

def generate_graph(lenarray, timeArray_best, time_arr_worst, time_arr_avg):
    x_values = lenarray
    y_best_values = timeArray_best
    y_worst_values = time_arr_worst
    y_avg_values = time_arr_avg

    plt.plot(x_values, y_best_values, label='Best Case', color='green', marker='.')
    plt.plot(x_values, y_worst_values, label='Worst Case', color='red', marker='.')
    plt.plot(x_values, y_avg_values, label='Average Case', color='blue', marker='.')

    plt.xlabel('Length of Array')
    plt.ylabel('Time of Execution')
    plt.title('Execution Time vs. Array Length')
    plt.grid()
    plt.legend()
    plt.show()

def analysis_best_case(arr):

    return timeit.timeit(lambda:binary_search(arr, arr[0]), number=1000)

def analysis_worst_case(arr):

    return timeit.timeit(lambda:binary_search(arr, arr[len(arr)- 1]), number=1000)

def analysis_avg_case(arr):

    return timeit.timeit(lambda:binary_search(arr, arr[random.randint(0,len(arr)-1)]), number= 1000)

if __name__ == '__main__':
    lengths = [ 60000,80000, 100000, 500000, 1000000, 2000000, 2500000]
    best_case = []
    worst_case = []
    avg_case = []

    for length in lengths:
        a = create_list(length)

        best = analysis_best_case(a)
        best_case.append(best)  # Extend the list with the results
        worst = analysis_worst_case(a)
        worst_case.append(worst)
        avg = analysis_avg_case(a)
        avg_case.append(avg)  # Extend the list with the results

    print(best_case)
    print(avg_case)
    print(worst_case)

    # print()
    with open("C:\\Users\\pragu\\OneDrive\\Desktop\\Book1.csv", 'w', newline='') as csvFile:
        csv_writer = csv.writer(csvFile)
        # Adding headings to the CSV file
        csv_writer.writerow(['Run', 'Array Length', 'Execution Time Best', 'Execution Time worst', 'Execution Time Avg'])
        for i in range(len(lengths)):
            csv_writer.writerow([i, lengths[i], best_case[i], worst_case[i], avg_case[i]])
    generate_graph(lengths , best_case, worst_case, avg_case)


    
        
```
